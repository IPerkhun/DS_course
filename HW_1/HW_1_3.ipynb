```python
data = scipy.random.exponential(scale=10, size=1000)
h0, h1 = 10, 15

def log_likelihood(theta0, theta1, x):
    L = (theta1 - theta0) / (theta0*theta1)*x
    return L


alpha, beta = 0.05, 0.05
a = np.log(beta / (1 - alpha))
b = np.log((1 - beta) / alpha)
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = {'x': [], 'y': []}
B = {'x': [], 'y': []}
Si = 0
x, y = [], []
flag = True

for i in range(1000):
    A['y'].append(a + np.log(h1 / h0)*i)
    B['y'].append(b + np.log(h1 / h0)*i)
    A['x'].append(i)
    B['x'].append(i)
    
i = 0
while flag == True:
    Si  += log_likelihood(h0, h1, data[i])
    x.append(i)
    y.append(Si)
    if (Si < B['y'][i]) and (Si > A['y'][i]):
        print(i + 1, 'continue')
        i +=1
        continue
    elif Si >= B['y'][i]:
        flag = False
        print(i + 1, 'Accept H1')
    elif Si <=  A['y'][i]:
        flag = False
        print(i + 1, 'Accept H0')
    
        
fig, ax = plt.subplots()
ax.plot(x, y, '.')
ax.plot(A['x'], A['y'], color = 'blue')
ax.plot(B['x'], B['y'], color = 'red')

legend = ax.legend()
plt.show()
```
