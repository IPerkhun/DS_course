```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
def get_PPS_sample(m1, m2, m3, s1, s2, s3, N, n):
    mix = []
    a = np.random.normal(loc= m1, scale=s1, size= int(0.15*N))
    b = np.random.normal(loc= m2, scale=s2, size= int(0.25*N))
    c = np.random.normal(loc= m3, scale=s3, size= int(0.6*N))
    
    
    std_a = np.std(a)
    std_b = np.std(b)
    std_c = np.std(c)
    
    C = sum([len(a) * std_a, len(b)* std_b, len(c)* std_c])

    n1 = (n*(len(a)*std_a)) / C
    n2 = (n*(len(b)*std_b)) / C
    n3 = (n*(len(c)*std_c)) / C
    

    mix.extend(np.random.choice(a, int(n1)))
    mix.extend(np.random.choice(b, int(n2)))
    mix.extend(np.random.choice(c, int(n3)))
    return np.array(mix)
```

```python
def get_PS_sample(m1, m2, m3, s1, s2, s3, N, n):
    mix = []
    a = np.random.normal(loc= m1, scale=s1, size= int(0.15*N))
    b = np.random.normal(loc= m2, scale=s2, size= int(0.25*N))
    c = np.random.normal(loc= m3, scale=s3, size= int(0.6*N))

    n1 = (len(a) * n) / N
    n2 = (len(b) * n) / N
    n3 = (len(c) * n) / N

    mix.extend(np.random.choice(a, int(n1)))
    mix.extend(np.random.choice(b, int(n2)))
    mix.extend(np.random.choice(c, int(n3)))
    return np.array(mix)
```

```python
def get_random_sample(m1, m2, m3, s1, s2, s3, N, n):
    mix = []
    a = np.random.normal(loc= m1, scale=s1, size= int(0.15*N))
    b = np.random.normal(loc= m2, scale=s2, size= int(0.25*N))
    c = np.random.normal(loc= m3, scale=s3, size= int(0.6*N))

    mix = np.concatenate([a, b, c])
    mix = np.random.choice(mix, n)
    
    return np.array(mix)
```

Одинаковое мат ожидание и дисперсия

```python
m1, m2, m3, s1, s2, s3,  = 100, 100, 100, 10, 10, 10
N = 500
n = 40
res = []
for i in range(1000):
    data = get_random_sample(m1, m2, m3, s1,s2,s3,N,n)
    res.append({
        'std': data.std(),
        'median': np.median(data),
        'var': data.var()
    })
    
res = pd.DataFrame(res)
res.describe()
```

```python
m1, m2, m3, s1, s2, s3,  = 100, 100, 100, 10, 10, 10
N = 40
n = 10
res = []
for i in range(1000):
    data = get_PPS_sample(m1, m2, m3, s1,s2,s3,N,n)
    res.append({
        'std': data.std(),
        'var': data.var()
    })
    
res = pd.DataFrame(res)
res.describe()
```
